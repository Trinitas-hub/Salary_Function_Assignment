```python
def get_employee_details(name, df):
    """
    Accepts an employee name and returns their details (row(s)) from the salary dataset.
    Handles errors like empty input and missing columns.
    """
    try:
        # validate input
        if not isinstance(name, str) or not name.strip():
            raise ValueError("Employee name must be a non-empty string.")

        # make sure column exists
        if "EmployeeName" not in df.columns:
            raise KeyError("EmployeeName column not found in the dataset.")

        name_clean = name.strip().lower()

        # find matches (case-insensitive)
        matches = df[df["EmployeeName"].astype(str).str.lower() == name_clean]

        if matches.empty:
            return f"No employee found with name: {name}"

        return matches

    except Exception as e:
        return f"Error: {e}"
```

```python
import os
import zipfile

def export_employee_to_zip(name, df):
    try:
        details = get_employee_details(name, df)

        if isinstance(details, str):
            return details  # returns error or "not found"

        # Create folder
        folder = "Employee Profile"
        os.makedirs(folder, exist_ok=True)

        # Save CSV
        safe_name = name.strip().replace(" ", "_")
        csv_path = os.path.join(folder, f"employee_details_{safe_name}.csv")
        details.to_csv(csv_path, index=False)

        # Zip folder
        zip_name = "Employee Profile.zip"
        with zipfile.ZipFile(zip_name, "w", zipfile.ZIP_DEFLATED) as z:
            for root, _, files in os.walk(folder):
                for file in files:
                    full_path = os.path.join(root, file)
                    z.write(full_path, arcname=os.path.relpath(full_path, start="."))

        return f"Saved CSV to {csv_path} and created {zip_name}"

    except Exception as e:
        return f"Export error: {e}"
```
